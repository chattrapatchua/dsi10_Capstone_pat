```python
# Import libraries and modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import np_utils

import h5py
import joblib
import radialProfile
import cv2
```

```python
def dff_img(img_array):
    psd1D_array = np.empty([img_array.shape[0], 158])
    for i, img in enumerate(img_array):
        # use numpy library to apply fourier transform
        f = np.fft.fft2(img.reshape(227,227))
        # shift areas of low frequency (by default at the top left corner) to the center 
        # areas of low frequency indicate that 
        fshift = np.fft.fftshift(f)
        magnitude_spectrum = 20*np.log(np.abs(fshift))
        psd1D = radialProfile.azimuthalAverage(magnitude_spectrum)
        psd1D_array[i,:] = psd1D
        #pd.DataFrame(psd1D_array).to_csv(f'../dataset/{img_array}_dff.csv', index=False)
    return psd1D_array
```

# Meta Classifier -  Voting Classifier

- Since the CNN does not seem to classify images with very thin cracks very well,
- and whereas the machine learning models are able to perform better on these types of cracks
- a voting classifier will be constructed
- ultimately, it was decided that 2 models would be used in the voting classifier: CNN and random forest
    - It was found that using all constructed models resulted in overfitting to Class 1 images. Classification on unseen images of Class 0 resulted in false positives most of the time.
    - Random forest model seemed to have the most reliable performance across both classes of images.
    - Furthermore, the pred_proba value of the CNN appeared to be a good indicator
        - i.e., if the pred_proba value was a very low number (i.e. < 0.02) then it was a strong indicator that the image was a negative class.

```python
def voting_classifier(cnn_model = [cnn_model2], 
                      ml_models = [rf_model]): #[logreg_model, rf_model, ada_model, xgb_model, knn_model, svm_model]): 
    pred_probas = []
    preds = None
    pred = None
    for model in cnn_model:
        pred_probas.append(model.predict_proba(img)[0][0])
    
    for model in ml_models:
        pred_probas.append(model.predict_proba(img_dff)[0][1])
    
    print(pred_probas)
    preds = [1 if proba >= 0.5 else 0 for proba in pred_probas]
    #print(preds)
    if (sum(preds) >= 1) & (pred_probas[0] > 0.02):
        pred = 1
        #print(f"{sum(preds)} out of {len(preds)} models: Surface has a crack.") 
        print("Conclusion: Cracked Surface (Class 1)")
    else:
        pred = 0
        #print(f"{sum(preds)} out of {len(preds)} models: Surface has a crack.") 
        print("Conclusion: Non cracked surface (Class 0)")
        
    return pred
  
```

# Conclusion

- Models constructed were able to detect obvious cracks with very high degree of accuracy
- CNN model unable to detect thin/small cracks, but machine learning models are quite good at this
- Machine learning models tend to misclassify Class 0 surfaces with certain characteristics (variations in color/pockmarked surfaces), whereas CNN model is relatively good at picking up these nuances
- Meta classifier (voting classifier) used to increase reliability

**Limitations**
- Generally, 2 scenarios that caused misclassification:
    - Thin, narrow cracks
    - Surfaces with a lot of small craters
- Cracks are only 1 type of defect
